```python
train, test = train_test_split(clean_data, test_size=0.15)
print(len(train))
print(len(test))

train.columns
```

```python
train_predictors = train.drop(['price', 'id'], axis=1)
train_target = train.drop(['id', 'bedrooms', 'bathrooms', 
'sqft_living', 'sqft_lot','floors', 'condition', 'grade', 'sqft_above', 
'yr_built', 'zipcode','lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1)

test_predicators = train.drop(['price', 'id'], axis=1)
test_target = train.drop(['id', 'bedrooms', 'bathrooms', 
'sqft_living', 'sqft_lot','floors', 'condition', 'grade', 'sqft_above', 
'yr_built', 'zipcode','lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1)



```

```python
print(train_predictors.head())
```

```python
lr = LinearRegression(normalize=True)
lr.fit(train_predictors, train_target)
train_predictions=lr.predict(train_predictors)
train_predictions
```

```python
plt.figure(figsize=(20,12))
plt.scatter(train_predictions, train_target)
plt.show()
print(metrics.r2_score(train_target, train_predictions))
```
